```python
import matplotlib.pyplot as plt
import numpy as np
import random

time=10
s12=0.1
s21=0.3
x1=0.3
x2=0.7
d=x2-x1
X1=[x1]
X2=[x2]
D=[d]
N=[0]
```

```python
for i in range(time):
    x1+=s12*d
    x2+=-s21*d
    d=x2-x1
    N+=[i]
    D+=[d]
    X1+=[x1]
    X2+=[x2]
```

```python
plt.figure()
plt.xlabel('Time')
plt.ylabel('X1 and X2')
plt.plot(N,X1)
plt.plot(N,X2)
plt.title('X1 (blue) and X2 (orange) in function of time')
plt.grid(True)
plt.show()
```

```python
def sensibility(age):
    return (1/3)*np.exp(-(age%100)/30)

t = np.arange(0, 300, 1)
plt.figure()
plt.xlabel('Time')
plt.ylabel('Sensibility')
plt.plot(t,sensibility(t))
plt.grid(True)
plt.show()
```

```python
'''CREATION OF A POPULATION'''

pop=1000 #size of the population
meetings=100 #number of meetings per year
time=5000 #total time of the study in years

POP=[]
for i in range (pop):
    POP+=[random.uniform(0, 1)]

TIME=[]
for i in range(time):
    TIME+=[i]

AGE=[]

for i in range(pop):
    r=random.randint(0,99)
    AGE+=[r]

'''DEATHS, BIRTHS, AND CHANGES OF OPINIONS'''

OPINIONS=np.zeros((time,pop))
OPINIONS[0]=POP

for t in range(1,time):
    for i in range(meetings):
        j=random.randint(0,pop-1)
        k=random.randint(0,pop-1)
        d=POP[k]-POP[j]
        if (abs(d)<0.10):
            POP[j]+=d*sensibility(AGE[j]+t)
            POP[k]-=d*sensibility(AGE[k]+t)
    OPINIONS[t]=POP
print(OPINIONS.T)
```

```python
plt.figure()
plt.xlabel('Time')
plt.ylabel('Opinions')
for i in range(pop):
    plt.plot(TIME,(OPINIONS.T)[i])
plt.title('Opinions in function of time')
plt.grid(True)
plt.show()
```
