```python
import matplotlib.pyplot as plt

n=100
s12=-0.1
s21=-0.1
x1=0.3
x2=0.7
d=x2-x1
X1=[x1]
X2=[x2]
D=[d]
N=[0]
```

```python
for i in range(n):
    x1+=s12*d
    x2+=-s21*d
    d=x2-x1
    N+=[i]
    D+=[d]
    X1+=[x1]
    X2+=[x2]
```

```python
plt.figure()
plt.xlabel('Time')
plt.ylabel('X1 and X2')
plt.plot(N,X1)
plt.plot(N,X2)
plt.title('X1 and X2 in function of time')
plt.grid(True)
plt.show()
```
